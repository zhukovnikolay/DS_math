#### 1. Найти область определения функции.
$$z = \sqrt{1-x^3} + \ln(y^2 - 1)$$

Видим два условия:
1. Подкоренное выражение $1 - x^3$  должно быть неотрицательно.
2. Выражение $y^2 - 1$, исходя из области определения функции натурального логарифма, должно быть положительно.

В результате получаем область определения исследуемой функции D(x; y), состоящую из объединения двух следующих систем:
\begin{equation*}
 \begin{cases}
   x \leq 1\\
   y > 1
 \end{cases}
\end{equation*}
____
\begin{equation*}
 \begin{cases}
   x \leq 1\\
   y < -1
 \end{cases}
\end{equation*}

```python
# графическое отображение области определения
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-2,1.1,0.1)
plt.axhline(y=0, lw=5, color='k')
plt.axvline(x=0, lw=5, color='k')
plt.axvline(x=1)
plt.axhline(y=1)
plt.axhline(y=-1)
plt.fill_between(x, -4, -1, color = "blue", alpha = 0.3)
plt.fill_between(x, 1, 4, color = "blue", alpha = 0.3)
plt.axis([-2, 3, -4, 4])
plt.grid()
```

```python
# график функции
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1, projection='3d')

x, y = np.meshgrid(np.linspace(-2, 2, 40), np.linspace(-4, 4, 80))
z = np.sqrt(1 - x**3) + np.log(y ** 2 - 1)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.plot_surface(x, y, z)

plt.show()
```

#### 4. Исследовать на экстремум функцию
$$z = x^2 + xy + y^2 - 6x - 9y$$

#### Необходимые условия экстремума:
\begin{equation*}
 \begin{cases}
   \displaystyle\frac{\partial z}{\partial x} = 2x + y - 6 = 0\\
   \displaystyle\frac{\partial z}{\partial y} = 2y + x - 9 = 0
 \end{cases}
\end{equation*}

Из (1): y = 6 - 2x

В (2): 12 - 4x + x - 9 = 0 => x = 1, y = 4

Необходимое условие выполнется. (1; 4) - стационарная точка.

#### Достаточные условия.
Вычислим вторые частные производные:

$$\frac{\partial^2 z}{\partial x^2} = 2, \frac{\partial^2 z}{\partial y^2} = 2, \frac{\partial^2 z}{\partial x\partial y} = 1$$

Подставим значения в матрицу Гессе и посчитаем ее определитель:
\begin{equation*}
\begin{vmatrix}
2 & 1 \\
1 & 2
\end{vmatrix}
= 3 > 0
\end{equation*}

Необходимое условие выполянется. При этом, знак второй частной производной по x > 0 => данная точка - точка минимума.

```python
# график функции
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1, projection='3d')

x, y = np.meshgrid(np.linspace(-4, 4, 40), np.linspace(-8, 8, 80))
z = x**2 + x*y + y**2 - 6*x - 9*y

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.scatter3D(1, 4, -21, c='r', s = 30)
ax.plot_surface(x, y, z)

plt.show()
```
