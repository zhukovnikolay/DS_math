#### 1. Исследовать функцию на условный экстремум

$$U = 3 - 8x + 6y, если\, x^2 + y^2 = 36$$

Функция Лагранжа:
$$L = 3 - 8x + 6y + \lambda\cdot(x^2 + y^2 - 36)$$

#### Необходимые условия экстремума:
$$\begin{equation*}
 \begin{cases}
   \displaystyle L'_x = -8 + 2\lambda x = 0\\
   \displaystyle L'_y = 6 + 2\lambda y = 0\\
   \displaystyle x^2 + y^2 - 36 = 0
 \end{cases}
 \Rightarrow
  \begin{cases}
   \displaystyle x = \frac{4}{\lambda}\\
   \displaystyle y = \frac{-3}{\lambda}\\
   \displaystyle \left(\frac{4}{\lambda}\right)^2 + \left(\frac{-3}{\lambda}\right)^2 = 36
 \end{cases}
 \Rightarrow
  \begin{cases}
   \displaystyle x = \frac{4}{\lambda}\\
   \displaystyle y = \frac{-3}{\lambda}\\
   \displaystyle \lambda^2 = \frac{25}{36}
 \end{cases} 
\end{equation*}
$$

Получаем две точки $\left(\frac{5}{6};4\frac{4}{5};-3\frac{3}{5}\right)$, $\left(-\frac{5}{6};-4\frac{4}{5};3\frac{3}{5}\right)$

#### Достаточные условия экстремума:
$$L''_{xx} = 2\lambda,\, L''{yy} = 2\lambda,\, L''{\lambda\lambda} = 0$$

$$L''{x\lambda} = 2x, \, L''{y\lambda} = 2y, \, L''{xy} = 0$$

\begin{equation*}
\begin{pmatrix}
L''_{\lambda\lambda} & L''_{\lambda x} & L''_{\lambda y}\\
L''_{x \lambda} & L''_{xx} & L''_{xy}\\
L''_{y \lambda} & L''_{yx} & L''_{yy}
\end{pmatrix}
= \begin{pmatrix}
0 & 2x & 2y\\
2x & 2\lambda & 0\\
2y & 0 & 2\lambda
\end{pmatrix}
\end{equation*}

\begin{equation*}
\begin{vmatrix}
0 & 2x & 2y\\
2x & 2\lambda & 0\\
2y & 0 & 2\lambda
\end{vmatrix}
= 0\cdot\begin{vmatrix}
2\lambda & 0\\
0 & 2\lambda
\end{vmatrix}
-2x\cdot\begin{vmatrix}
2x & 0\\
2y & 2\lambda
\end{vmatrix}
+2y\cdot\begin{vmatrix}
2x & 2\lambda\\
2y & 0
\end{vmatrix}
= 0 - 8x^2\lambda - 8y^2\lambda = -8\lambda(x^2 + y^2)
\end{equation*}

Таким образом, знак определителя больше нуля при $\lambda < 0$ и наоборот => $\left(\frac{5}{6};4\frac{4}{5};-3\frac{3}{5}\right)$ - точка минимума, $\left(-\frac{5}{6};-4\frac{4}{5};3\frac{3}{5}\right)$ - точка максимума

#### 3. Найти производную функции $U = x^2 + y^2 + z^2$ по направлению вектора $\vec{c}=(-9,8,-12)$ в точке $M(8;-12;9)$

Нормализуем вектор $\vec{c}$:
$$|\vec{c}| = \sqrt{x_0^2 + y_0^2 + z_0^2} = \sqrt{(-9)^2 + 8^2 + (-12)^2} = 17$$

$$\vec{c_0} = \frac{\vec{c}}{|\vec{c}|} = \left(-\frac{9}{17},\frac{8}{17},-\frac{12}{17}\right)$$

Вычислим частные производные функции U:
$$U'_x = 2x, \, U'_y = 2y, \, U'_z = 2z$$

Найдем градиент:
$$grad\,U|_{(8,-12,9)} = (16, -24, 18)$$

Найдем производную по направлению вектора:
$$U'_{\vec{c}}|_{(8,-12,9)} = -16\cdot\frac{9}{17} - 24\cdot\frac{8}{17} - 18\cdot\frac{12}{17} = -\frac{552}{17}$$

Исследуемая функция убывает с вычисленной скоростью.

#### 4. Найти производную функции $U = e^{x^2 + y^2 + z^2}$ по направлению вектора $\vec{d}=(4,-13,-16)$ в точке $L(-16;4;-13)$

Нормализуем вектор $\vec{d}$:
$$|\vec{d}| = \sqrt{x_0^2 + y_0^2 + z_0^2} = \sqrt{4^2 + (-13)^2 + (-16)^2} = 21$$

$$\vec{d_0} = \frac{\vec{d}}{|\vec{d}|} = \left(\frac{4}{21},-\frac{13}{21},-\frac{16}{21}\right)$$

Вычислим частные производные функции U:
$$U'_x = 2xe^{x^2 + y^2 + z^2}, \, U'_y = 2ye^{x^2 + y^2 + z^2}, \, U'_z = 2ze^{x^2 + y^2 + z^2}$$

Найдем градиент:
$$grad\,U|_{(-16,4,-13)} = (-32e^{441}, 8e^{441}, -26e^{441})$$

Найдем производную по направлению вектора:
$$U'_{\vec{d}}|_{(-16,4,-13)} = -32e^{441}\cdot\frac{4}{21} - 8e^{441}\cdot\frac{13}{21} + 26e^{441}\cdot\frac{16}{21} = \frac{184e^{441}}{21}$$

Исследуемая функция возрастает с вычисленной скоростью.
