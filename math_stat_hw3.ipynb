**1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.**

```python
# упорядочим выборку программно
import numpy as np
z = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
z.sort()
z
```

```python
z.shape[0]
```

Вычисляем:
1. Среднее арифметическое: $$\mu = \overline x = \displaystyle\frac{17+24+25+30+33+45+55+57+65+65+70+75+75+77+80+84+89+90+100+150}{20} = 65.3$$


2. Смещенная оценка дисперсии: $$s^2 = \displaystyle\frac{\sum_{i=1}^{20}(x_i-\overline x)^2}{20} = 950.11$$


3. Несмещенная оценка дисперсии: $$s^2 = \displaystyle\frac{\sum_{i=1}^{20}(x_i-\overline x)^2}{19} = 1000.116$$


4. Среднее квадратичное отклонение: $$s = \sqrt{s^2} = 31.62$$

```python
# вычисляем программно без функций
mu = z.sum()/z.shape[0]
print(f'Матожидание: {mu}')
z_var = pow(z - mu, 2)
print(f'Смещенная оценка дисперсии: {z_var.sum()/z_var.shape[0]}')
print(f'Несмещенная оценка дисперсии: {z_var.sum()/(z_var.shape[0] - 1)}')
print(f'Среднее квадратичное отклонение: {np.sqrt(z_var.sum()/(z_var.shape[0] - 1))}')
```

```python
# проверяем
print(f'Матожидание: {z.mean()}')
print(f'Смещенная оценка дисперсии: {z.var()}')
print(f'Несмещенная оценка дисперсии: {z.var(ddof=1)}')
print(f'Среднее квадратичное отклонение: {z.std(ddof=1)}')
```

**2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?**

Существует три подходящих события:
1. Из первого ящика достали 0 белых мячей, из второго - три.
2. Из первого ящика достали один белый мяч, из второго - два.
3. Из первого ящика достали два белых мяча, из второго - один.

Решим через сочетания:

1. Число благоприятствующих исходов для первого события: для ящика 1 - $C^0_{5} * C^2_{3}$, число всех возможных исходов: $C^{2}_{8}$, для ящика 2 - $C^3_{5} * C^1_{7}$, число всех возможных исходов: $C^{4}_{12}$

2. Число благоприятствующих исходов для второго события: для ящика 1 - $C^1_{5} * C^1_{3}$, число всех возможных исходов: $C^{2}_{8}$, для ящика 2 - $C^2_{5} * C^2_{7}$, число всех возможных исходов: $C^{4}_{12}$ 

3. Число благоприятствующих исходов для третьего события: для ящика 1 - $C^2_{5} * C^0_{3}$, число всех возможных исходов: $C^{2}_{8}$, для ящика 2 - $C^1_{5} * C^3_{7}$, число всех возможных исходов: $C^{4}_{12}$ 

Сумма вероятностей трех событий (т.к. они несовместны) будет искомой вероятностью.

```python
def combinations(n, k):
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))

p1 = ((combinations(5, 0) * combinations(3, 2))/combinations(8, 2)) * (combinations(5, 3) * combinations(7, 1))/combinations(12, 4)
print(f'Вероятность достать 0 белых мячей из первого ящика и 3 из второго: {p1}')
p2 = ((combinations(5, 1) * combinations(3, 1))/combinations(8, 2)) * (combinations(5, 2) * combinations(7, 2))/combinations(12, 4)
print(f'Вероятность достать 1 белый мяч из первого ящика и 2 из второго: {p2}')
p3 = ((combinations(5, 2) * combinations(3, 0))/combinations(8, 2)) * (combinations(5, 1) * combinations(7, 3))/combinations(12, 4)
print(f'Вероятность достать 2 белых мяча из первого ящика и 1 из второго: {p3}')
print(f'Искомая вероятность: {p1+p2+p3}')
```

**3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a) первым спортсменом б) вторым спортсменом в) третьим спортсменом.**

Обозначим события:
1. A - мишень поражена.
2. B1 - выстрел произвел первый спортсмен.
3. B2 - выстрел произвел второй спортсмен.
4. B3 - выстрел произвел третий спортсмен.

Тогда по формуле полной вероятности:
$$P(A) = \frac{1}{3}\cdot(0.9 + 0.8 + 0.6) = \frac{23}{30}$$

Определим искомые вероятности по формуле Байеса:
1) Вероятность того, что выстрел произведен первым спортсменом (из условия известно, что P(A|B1) = 0.9):
$$P(B1|A) = \displaystyle\frac{P(B)\cdot P(A|B1)}{P(A)} = \frac{\frac{1}{3}\cdot \frac{9}{10}}{\frac{23}{30}} = \frac{9}{23}$$

2) Вероятность того, что выстрел произведен вторым спортсменом (из условия известно, что P(A|B2) = 0.8):
$$P(B2|A) = \displaystyle\frac{P(B)\cdot P(A|B2)}{P(A)} = \frac{\frac{1}{3}\cdot \frac{8}{10}}{\frac{23}{30}} = \frac{8}{23}$$

3) Вероятность того, что выстрел произведен третьим спортсменом (из условия известно, что P(A|B3) = 0.6):
$$P(B3|A) = \displaystyle\frac{P(B)\cdot P(A|B3)}{P(A)} = \frac{\frac{1}{3}\cdot \frac{6}{10}}{\frac{23}{30}} = \frac{6}{23}$$

**4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a) на факультете A б) на факультете B в) на факультете C?**

Обозначим события:
1. A - студент сдал первую сессию.
2. B1 - студент учится на факультете A.
3. B2 - студент учится на факультете B.
4. B3 - студент учится на факультете C.

По формуле полной вероятности (из условия - вероятность того, что студент учится на факультете C - в два раза выше, чем на A или B):
$$P(A) = \frac{1}{4}\cdot0.8 + \frac{1}{4}\cdot0.7 + \frac{1}{2}\cdot 0.9 = \frac{33}{40}$$

Определим искомые вероятности по формуле Байеса:
1) Вероятность того, что сессию сдал студент с факультета A (из условия известно, что P(A|B1) = 0.8):
$$P(B1|A) = \displaystyle\frac{P(B)\cdot P(A|B1)}{P(A)} = \frac{\frac{1}{4}\cdot \frac{8}{10}}{\frac{33}{40}} = \frac{8}{33}$$

2) Вероятность того, что сессию сдал студент с факультета B (из условия известно, что P(A|B2) = 0.7):
$$P(B2|A) = \displaystyle\frac{P(B)\cdot P(A|B2)}{P(A)} = \frac{\frac{1}{4}\cdot \frac{7}{10}}{\frac{33}{40}} = \frac{7}{33}$$

3) Вероятность того, что сессию сдал студент с факультета C (из условия известно, что P(A|B3) = 0.9):
$$P(B3|A) = \displaystyle\frac{P(B)\cdot P(A|B3)}{P(A)} = \frac{\frac{1}{2}\cdot \frac{9}{10}}{\frac{33}{40}} = \frac{18}{33}$$

**5. Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а) все детали б) только две детали в) хотя бы одна деталь г) от одной до двух деталей?**

Обозначим события:
1. A - деталь 1 вышла из строя.
2. B - деталь 2 вышла из строя.
3. С - деталь 3 вышла из строя.

Выход из строя деталей - независимые события. Тогда:
1. Вероятность выхода из строя всех деталей в первый месяц:
$$P(A\cap B \cap C) = 0.1 \cdot 0.2 \cdot 0.25 = 0.005$$

2. Вероятность выхода из строя двух деталей в первый месяц можно найти как сумму вероятностей следующих несовместных событий:
* детали 1 и 2 вышли из строя, деталь 3 работает;
* детали 1 и 3 вышли из строя, деталь 2 работает;
* детали 2 и 3 вышли из строя, деталь 1 работает.
$$P = P(A\cap B \cap \overline C) + P(A\cap \overline B \cap C) + P(\overline A\cap B \cap C) = 0.1 \cdot 0.2 \cdot 0.75 + 0.1 \cdot 0.8 \cdot 0.25 + 0.9 \cdot 0.2 \cdot 0.25 = 0.015 + 0.02 + 0.045 = 0.08$$

3. Вероятность выхода из строя хотя бы одной детали в первый месяц можно найти как сумму вероятностей следующих несовместных событий:
* деталь 1 вышла из строя, детали 2 и 3 работают;
* детали 2 вышла из строя, детали 1 и 3 работают;
* деталь 3 вышла из строя, детали 1 и 2 работают.
$$P = P(A\cap \overline B \cap \overline C) + P(\overline A\cap B \cap \overline C) + P(\overline A\cap \overline B \cap C) = 0.1 \cdot 0.8 \cdot 0.75 + 0.9 \cdot 0.2 \cdot 0.75 + 0.9 \cdot 0.8 \cdot 0.25 = 0.06 + 0.135 + 0.18 = 0.375$$

4. Вероятность выхода из строя от одной до двух деталей в первый месяц можно найти как сумму вероятностей несовместных событий 2 и 3:
$$P = 0.08 + 0.375 = 0.455$$
