### Непрерывные случайные величины. Функция распределения и плотность распределения вероятностей. Равномерное и нормальное распределение. Центральная предельная теорема.

**1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.**

Среднее значение для равномерно распределенной случайной величины можно найти по формуле:
$$M(A) = \frac{a + b}{2} = \frac{200 + 800}{2} = 500$$

Дисперсия может быть найдена по формуле:
$$D(A) = \frac{(b-a)^2}{12} = \frac{(800 - 200)^2}{12} = 30000$$

**2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.**

Из формулы нахождения дисперсии равномерно распределенной случайной величины:
$$D(B) = \frac{(b-a)^2}{12} \Rightarrow b = \sqrt{12D(B)} + a = \sqrt{2.4} + 0.5 \approx 2.049$$

Среднее в таком случае равно:
$$M(B) = \frac{a + b}{2} \approx \frac{0.5 + 2.049}{2} \approx 1.2745$$

**3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения
f(x) = (1 / (4 * sqrt(2*pi))) * (exp(-((x+2)^2)/32)). Найдите: M(X), D(X), std(X)**

Из формулы функции плотности вероятности:
$$f(x) = \displaystyle\frac{1}{\sigma\sqrt{2\pi}}\cdot \displaystyle e^{-\displaystyle\frac{(x-\mu)^2}{2\sigma^2}}$$

Таким образом, $M(X) = -2$, $D(X) = 16$, $std(X) = 4$

Таким образом, исследуемая случайная величина $X \sim N(174, 64)$.

$$Z = \frac{182 - 174}{8} = 1$$

Искомая вероятность в таком случае составит 1 - P = 0.15866

$$Z = \frac{190 - 174}{8} = 2$$
Это соответствует табличному значению P = 0.97725.

Искомая вероятность в таком случае составит 1 - P = 0.02275

Решим через Z-таблицы.
$$Z = \frac{166 - 174}{8} = -1$$
Это соответствует табличному значению P1 = 0.15866 (вероятность того, что случайный человек ниже 166 сантиметров). Из части б) учитываем вероятность P2 = 0.97725 (вероятность того, что случайный человек ниже 190 сантиметров)

Искомая вероятность в таком случае составит P2 - P1 = 0.81859

**г) от 166 см до 182 см**

Из части а) и в) знаем вероятности людей оказаться ниже 166 см (P1 = 0.15866) и выше 182 см (P2 = 0.15866).
Искомая вероятность в таком случае составит 1 - (P1 + P2) = 0.68268

**д) от 158 см до 190 см**

$$Z = \frac{158 - 174}{8} = -2$$
Это соответствует табличному значению P1 = 0.02275 (вероятность того, что случайный человек ниже 158 сантиметров). Из части б) учитываем вероятность P2 = 0.02275 (вероятность того, что случайный человек выше 190 сантиметров)

**5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?**

Случайная величина $X \sim N(178, 25)$ имеет среднее квадратичное отклонение = 5. 
Для исследуемой величины вычислим число z, которое и будет являться искомой величиной.
$$z = \frac{190 - 178}{5} = 2.4$$
