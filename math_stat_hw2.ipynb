### Дискретные случайные величины. Закон распределения вероятностей. Биномиальный закон распределения. Распределение Пуассона

**1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.**

Случайная величина (попадания в цель) подчиняется биномиальному распределению, т.к. нам задана точная вероятность, число испытаний конечно и невелико. По формуле Бернулли:
$$P(k=85) = C_n^k\cdot p^k\cdot q^{n-k} = C_{100}^{85}\cdot 0.8^{85}\cdot 0.2^{15} = \frac{100!}{85!(100-85)!}\cdot 0.8^{85}\cdot 0.2^{15} = 0.048$$

```python
# проверим программно
import numpy as np
def combinations(n, k):
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))
k = 85
n = 100
p = 0.8
combinations(n, k) * p**k * (1 - p)**(n - k)
```

**2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?**

Случайная величина (перегоревшие лампочки) подчиняется распределению Пуассона, т.к. вероятностьт наступления события мала, необходимо найти вероятность наступления событий в единицу измерения (день).

По формуле Пуассона ($\lambda = 5000 * 0.0004 = 2$), вероятность того, что не перегорит ни одна:
$$P(0) \approx \frac{\lambda^m}{m!}\cdot e^{-\lambda} \approx \frac{2^0}{0!}\cdot e^{-2} \approx 0.1353$$

Вероятность того, что перегорят ровно две:
$$P(2) \approx \frac{2^2}{2!}\cdot e^{-2} \approx 0.27067$$

```python
# проверим программно
m1 = 0
m2 = 2
p = 0.0004
n = 5000
print(f'Вероятность того, что не перегорит ни одна: {((p * n)**m1/np.math.factorial(m1))*np.math.exp(-p * n)}')
print(f'Вероятность того, что перегорит две: {((p * n)**m2/np.math.factorial(m2))*np.math.exp(-p * n)}')
```

**3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?**

Случайная величина (выпадение орла) подчиняется биномиальному распределению, т.к. нам известна точная вероятность события (0.5), число испытаний конечно и невелико, необходимо найти вероятность события k раз из n. По формуле Бернулли:
$$P(k=70) = C_n^k\cdot p^k\cdot q^{n-k} = C_{144}^{70}\cdot 0.5^{70}\cdot 0.5^{74} = \frac{144!}{70!(144-70)!}\cdot 0.5^{70}\cdot 0.5^{74} = 0.0628$$

```python
# проверим программно
k = 70
n = 144
p = 0.5
combinations(n, k) * p**k * (1 - p)**(n - k)
```

**4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые?**

Возможно два решения -  через сочетания (все элементы множества участвуют, порядок неважен) и через пару независимых событий, состоящие из зависимых событий. Решим обоими способами. 
Обозначим события:
* A - первый извлеченный мяч из первого ящика - белый.
* B - второй извлеченный мяч из первого ящика - белый.
* C - первый извлеченный мяч из второго ящика - белый.
* D - второй извлеченный мяч из второго ящика - белый.

События A и B, а также C и D зависимы, т.к. наступление каждого из них меняет вероятность наступления другого. При этом пары событий независимы, т.к. происходят отдельно друг от друга. Тогда из формулы вероятности появления зависимых/независимых событий:
$$P((A*B)(C*D)) = (P(A)*P(B|A))*(P(C)*P(D|C)) = \frac{7}{10}\cdot\frac{6}{9}\cdot\frac{9}{11}\cdot\frac{8}{10} = 0.30(54)$$

____
Решим программно через сочетания:

Число благоприятствующих исходов для первого ящика: $C^2_{7} * C^0_{3}$, число всех возможных исходов: $C^{2}_{10}$. Их отношение будет искомой вероятностью для первого ящика.

Число благоприятствующих исходов для второго ящика: $C^2_{9} * C^0_{2}$, число всех возможных исходов: $C^{2}_{11}$. Их отношение будет искомой вероятностью для второго ящика.

Произведление первого и второго результата будет искомой вероятностью.

```python
p1 = (combinations(7, 2) * combinations(3, 0))/combinations(10, 2)
print(f'p1 = {p1}')
p2 = (combinations(9, 2) * combinations(2, 0))/combinations(11, 2)
print(f'p2 = {p2}')
print(f'Искомая вероятность: {p1*p2}')
```

**В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что ровно два мяча белые?**

Данное событие может произойти в одном из трех случаев:
1. Из первого и второго ящика достали по одному белому мячу (сумма вероятностей достать сначала белый и потом черный, и наоборот).
2. Из первого ящика достали два белых мяча, из второго - ни одного.
3. Из первого ящика не достали ни одного белого мяча, из второго - два.

Найдем вероятность каждого из них:

$$P1 = \left(\frac{7}{10}\cdot\frac{3}{9} + \frac{3}{10}\cdot\frac{7}{9}\right)\cdot\left(\frac{9}{11}\cdot\frac{2}{10} + \frac{2}{11}\cdot\frac{9}{10}\right) = 0.15(27)$$

$$P2 = \frac{7}{10}\cdot\frac{6}{9}\cdot\frac{2}{11}\cdot\frac{1}{10} = 0.008(48)$$

$$P3 = \frac{3}{10}\cdot\frac{2}{9}\cdot\frac{9}{11}\cdot\frac{8}{10} = 0.043(63)$$

Искомая вероятность будет суммой вероятностей перечисленных событий, т.к. они несовместны. 
$$P = P1 + P2 + P3 = 0.20(48)$$

Решим программно через сочетания:

Число благоприятствующих исходов для первого события: для ящика 1 - $C^1_{7} * C^1_{3}$, число всех возможных исходов: $C^{2}_{10}$, для ящика 2 - $C^1_{9} * C^1_{2}$, число всех возможных исходов: $C^{2}_{11}$ 

Число благоприятствующих исходов для второго события: для ящика 1 - $C^2_{7} * C^0_{3}$, число всех возможных исходов: $C^{2}_{10}$, для ящика 2 - $C^0_{9} * C^2_{2}$, число всех возможных исходов: $C^{2}_{11}$ 

Число благоприятствующих исходов для третьего события: для ящика 1 - $C^0_{7} * C^2_{3}$, число всех возможных исходов: $C^{2}_{10}$, для ящика 2 - $C^2_{9} * C^0_{2}$, число всех возможных исходов: $C^{2}_{11}$ 

Сумма вероятностей трех событий будет искомой вероятностью.

```python
p1 = ((combinations(7, 1) * combinations(3, 1))/combinations(10, 2)) * ((combinations(9, 1) * combinations(2, 1))/combinations(11, 2))
print(f'Вероятность достать по одному белому мячу из каждого ящика: {p1}')
p2 = ((combinations(7, 2) * combinations(3, 0))/combinations(10, 2)) * (combinations(9, 0) * combinations(2, 2))/combinations(11, 2)
print(f'Вероятность достать два белых мяча из первого ящика и 0 из второго: {p2}')
p3 = ((combinations(7, 0) * combinations(3, 2))/combinations(10, 2)) * (combinations(9, 2) * combinations(2, 0))/combinations(11, 2)
print(f'Вероятность достать два белых мяча из второго ящика и 0 из первого: {p3}')
print(f'Искомая вероятность: {p1+p2+p3}')
```

**В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что хотя бы один мяч белый?**

Благоприятным исходом будут являться все события, при которых хотя бы один вытащенный мяч будет белым. Тогда найдем вероятность того, что все вытащенные мячу будут черными, в таком случае искомая вероятность будет равняться 1 - p, где p - вероятность того, что все вытащенные мячи - черные. По аналогии с первой частью задачи 4:
$$P((A*B)(C*D)) = (P(A)*P(B|A))*(P(C)*P(D|C)) = \frac{3}{10}\cdot\frac{2}{9}\cdot\frac{2}{11}\cdot\frac{1}{10} = 0.00(12)$$

Искомая вероятность:
$$P = 1 - 0.00(12) = 0.99(87)$$
